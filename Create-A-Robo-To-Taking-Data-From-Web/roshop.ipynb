```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

options = Options()
options.add_experimental_option("detach",True)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), 
                          options=options)
driver.maximize_window()
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:\\Users\\1230217\\OneDrive - ELOBR\\Shopping_Abrasce_links.csv", sep=';', encoding='utf-8')
```

```python
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

#driver.get('https://www3.tjrj.jus.br/consultaprocessual/api/processos/por-cpf-cnpj')
driver.get('https://www3.tjrj.jus.br/consultaprocessual/#/consultapublica#porNumero')
```

```python
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

# Processo de input e coleta
for i, link in enumerate(df['link']):
    print(i+1,"/613")
    try:
        time.sleep(1)
        driver.get(link)
        time.sleep(5)
        
        administradora = WebDriverWait(driver,10).until(EC.visibility_of_element_located((By.XPATH,"/html/body/div[4]/div/div[3]/div/div[1]/div[2]/div[1]/div[1]/div/div/p/span"))).text
        area_bruta =  WebDriverWait(driver,10).until(EC.visibility_of_element_located((By.XPATH,"/html/body/div[4]/div/div[3]/div/div[1]/div[2]/div[2]/div/p[3]/span"))).text
        endereco = WebDriverWait(driver,10).until(EC.visibility_of_element_located((By.XPATH,"/html/body/div[4]/div/div[3]/div/div[2]/div/ul/li[3]/span[2]"))).text
    
        df.loc[i,"adm"] = administradora
        df.loc[i,"area_locavel"] = area_bruta
        df.loc[i,"endereco"] = endereco
        time.sleep(1)
    except:
        pass

df.head()
```
