**Importando bibliotecas necessárias**

```python
import pyautogui
import time
import pyperclip
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import urllib
```

```python
df = pd.read_excel("C:\\Users\\Tuta\\Desktop\\cnae_mcc.xlsx")
display(df)
```

**Robo coletador de dados**

**ACQIO**

```python
# Definição do site para consulta
navegador = webdriver.Chrome()
navegador.get("https://acqio.com.br/mcc/")

# Cordenada para input
loc = navegador.find_element_by_xpath('/html/body/section/div/div[2]/div[1]/form/div/input')

# Cordenada para reiniciar
fecha = navegador.find_element_by_xpath('/html/body/div[1]/div/div/div[2]/button')

# Tempo de espera
time.sleep(5)

# Processo de input e coleta
for i, codigo in enumerate(df['codigo']):
    try:
        loc.clear()
        time.sleep(1)
        loc.send_keys(codigo)
        time.sleep(1)
        loc.send_keys(Keys.ENTER)
        time.sleep(5)
        fecha.click()
        time.sleep(1)
        #loc.send_keys(Keys.ENTER)
        mcc_num = navegador.find_element_by_xpath('/html/body/section/div/div[2]/div[2]/ul/li[3]').text
        mcc_des = navegador.find_element_by_xpath('/html/body/section/div/div[2]/div[2]/ul/li[4]').text
        df.loc[i,"mcc"] = mcc_num
        df.loc[i,"mcc_descricao"] = mcc_des
    except:
        pass
```

```python
#Cria novas colunas para nova coleta em outra fonte
df["mcc_2"] = np.nan
df["mcc_2_descricao"] = np.nan
df.to_csv('C:\\Users\\Tuta\\Desktop\\cnae_mcc.csv', sep='\t', encoding='UTF-8')
display(df)
```
